# Number of matches played in each season.

```python
no_of_matches_eachseason=ipl_mt['year'].value_counts()
no_of_matches_eachseason=no_of_matches_eachseason.sort_index()
```

```python
Year=no_of_matches_eachseason.index
count=no_of_matches_eachseason.values
```

```python
colors = [ "#8BC34A", "#D4E157", "#FFB300", "#FF7043", "#CC113C", "#00AA5A" ]

plt.figure(dpi=100,figsize=(6,4))
plt.bar(Year,count,color=colors,edgecolor='black',alpha=0.7)
plt.xticks(rotation=90)
plt.xlabel('Year',fontsize=12)
plt.ylabel('No of matches'.title(),fontsize='12')
plt.title('Matches played in each season'.title())
for index, value in enumerate(count):
    plt.text(index-0.4,value+1, str(value))
plt.ylim(0,100)
plt.grid(alpha=0.2)
sns.set()
#sns.color_palette("rocket")
```

# Relationship between toss and winning

```python
toss_and_match=ipl_mt[ipl_mt['toss_winner']==ipl_mt['winner']]['winner'].value_counts()
tw=toss_and_match.to_frame()
print(tw)
toss_and_match
losetoss_and_winmatch=ipl_mt[ipl_mt['toss_winner']!=ipl_mt['winner']]['winner'].value_counts()
losetoss_and_winmatch
```

```python
fig=plt.figure(figsize=(14,5))
colors = [ "#8BC34A", "#D4E157", "#FFB300", "#FF7043", "#CC113C", "#00AA5A" ]
colors2 = [ "#8BC36A", "#D4E187", "#FFB700", "#FF7033", "#CC163C", "#00AA8A" ]

ax1=fig.add_subplot(1,2,1)
ax1.bar(tw.index,tw['winner'],color=colors,alpha=0.6)
ax1.set_ylabel(('No of matches won').title(),c='k',fontsize=12)
ax1.set_title(('won the toss and won the match').title(),fontsize=14,color='maroon')
ax1.grid()
ax1.set_facecolor('white')
ax2=fig.add_subplot(1,2,2)
ax2.bar(losetoss_and_winmatch.index,losetoss_and_winmatch.values,color=colors2)
ax1.tick_params(labelrotation=90)
ax2.tick_params(labelrotation=90)
ax2.set_title(('lost the toss and won the match').title(),c='b',fontsize=14)
ax2.set_yticks([])
ax2.grid()
ax2.set_facecolor('white')
for ax in [ax1, ax2]:
    for sp in ['top', 'bottom', 'left', 'right']:
        ax.spines[sp].set_visible(False)

for ind,val in enumerate(tw['winner']):
    ax1.text(ind-0.4,val+1,str(val))
for ind,val in enumerate(losetoss_and_winmatch.values):
    ax2.text(ind-0.4,val+1,str(val))                  
```

```python
yearwise_toss_decision=ipl_mt.groupby('year')['toss_decision'].value_counts()
```

```python
yearwise_toss_decisiondf.unstack().plot(kind='bar',figsize=(10,5),\
                                        title='Team choose to bat or field after winning toss')
#plt.figure(figsize=(6,6))
plt.legend(['bat','field'])
plt.show()
```

# Team won by chasing or defending?

```python
runs_wicket=ipl_mt.groupby('winner')['result'].value_counts()
runs_wicket=runs_wicket.to_frame()
uns=runs_wicket.unstack()

uns.plot(kind='bar',figsize=(20,10),fontsize=20,title='won by runs or wickets')

plt.legend(['runs','tie','wickets'],loc='upper left',fontsize=16)

plt.show()
```
